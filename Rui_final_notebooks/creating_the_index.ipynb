```python
# Ensure all text is in the same case and spaces are stripped
df_monetary_policy['Classification Joaquin'] = df_monetary_policy['Classification Joaquin'].str.lower().str.strip()
df_monetary_policy['Classification Ed'] = df_monetary_policy['Classification Ed'].str.lower().str.strip()
df_monetary_policy['Classification Rui'] = df_monetary_policy['Classification Rui'].str.lower().str.strip()

# Define the mapping
sentiment_mapping = {
    'dovish': -1,
    'neutral': 0,
    'hawkish': 1
}

# Apply the mapping to each column
df_monetary_policy['Classification Joaquin'] = df_monetary_policy['Classification Joaquin'].map(sentiment_mapping)
df_monetary_policy['Classification Ed'] = df_monetary_policy['Classification Ed'].map(sentiment_mapping)
df_monetary_policy['Classification Rui'] = df_monetary_policy['Classification Rui'].map(sentiment_mapping)

# Check the results
print(df_monetary_policy[['Classification Joaquin', 'Classification Ed', 'Classification Rui']].head())
```

```python
# Calculate the average of the three columns
df_monetary_policy['Average_Classification'] = df_monetary_policy[['Classification Joaquin', 'Classification Ed', 'Classification Rui']].mean(axis=1)

# Check the results
print(df_monetary_policy[['Classification Joaquin', 'Classification Ed', 'Classification Rui', 'Average_Classification']].head())
```

```python
#Add a column saying Name_of_Speaker and fill it with the name of the speaker
df_monetary_policy['Name_of_Speaker'] = 'ECB_MONETARY_POLICY_STATEMENT'
```

```python
df_monetary_policy.drop(['Classification Joaquin', 'Classification Ed', 'Classification Rui'], axis=1, inplace=True)
```

```python
df_monetary_policy
```

```python
df_monetary_policy=df_monetary_policy[['Date', 'Article', 'Name_of_Speaker', 'Average_Classification']]
```

```python
df_monetary_policy.rename(columns={'Article': 'Statement'}, inplace=True)
df_monetary_policy.rename(columns={'Average_Classification': 'Sentiment'}, inplace=True)
df_monetary_policy
```
